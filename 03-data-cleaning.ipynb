# Data Preparation


## Data Load

```python
import pandas as pd
import numpy as np
```

```python
file_in = "df_amazon_original.csv"
file_out = "df_processed.csv"
```

```python
df = pd.read_csv(file_in)
```

```python
df.head()
```

## Processing

```python
df.replace({None: np.nan}, inplace=True)
# df = df.set_index("").reset_index(drop=True).drop('_id',axis=1)
```

```python
df.info()
```

```python
# Null values
df.isna().sum()
```

```python
df[df.isna().any(axis=1)]
```

```python
# Drop Na
df = df.dropna()
```

```python
# Check for Duplicates
df.duplicated().sum()
```

## Change data type

```python
df.info()
```

```python
# Date to datetime
df['Time'] = pd.to_datetime(df['Time'], unit='s')
# Object to String
df['Summary'] = df['Summary'].astype('string')
df['Text'] = df['Text'].astype('string')
```

```python
df['Year'] = df['Time'].dt.year
df['Month'] = df['Time'].dt.month
df['Day'] = df['Time'].dt.day
```

```python
df.info()
```

## Data out

```python
df.to_csv(file_out)
```
